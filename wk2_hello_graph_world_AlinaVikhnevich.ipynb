```python
import networkx as nx
import matplotlib.pyplot as plt
```

### Create the graph
I am using the built-in Krackhardt Kite graph available in NetworkX. It is a small social network commonly used as a demo dataset.


```python
G = nx.krackhardt_kite_graph()
G
```

### Inspect nodes and edges
I am listing nodes and edges to confirm the structure loaded correctly.


```python
list(G.nodes())
```

```python
list(G.edges())[:15], G.number_of_edges()
```

### Basic summary statistics
These are quick checks that the graph is not empty and has reasonable structure.

```python
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())
print("Density:", nx.density(G))
```

### Degree centrality (example metric)
This gives a simple measure of how connected each node is relative to the maximum possible degree.

```python
deg_cent = nx.degree_centrality(G)
sorted(deg_cent.items(), key=lambda x: x[1], reverse=True)[:5]
```

### Visualize the network
I use a spring layout to make the structure easy to see.

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42)

nx.draw_networkx_nodes(G, pos, node_size=600)
nx.draw_networkx_edges(G, pos, width=1.2, alpha=0.8)
nx.draw_networkx_labels(G, pos, font_size=10)

plt.title("Krackhardt Kite Graph")
plt.axis("off")
plt.show()
```
