### Degree centrality (example metric)
Degree centrality provides a simple measure of how connected each node is relative to the maximum possible number of connections in the network. Nodes with higher degree centrality occupy more structurally prominent positions, as they maintain direct ties to a larger number of other nodes. This metric offers an intuitive first look at potential influence and visibility within the network.

```python
deg_cent = nx.degree_centrality(G)
sorted(deg_cent.items(), key=lambda x: x[1], reverse=True)[:5]
```

### Visualize the network
I use a spring layout to make the structure easy to see.

```python
plt.figure(figsize=(8, 6))

pos = nx.spring_layout(G_named, seed=42)

nx.draw_networkx_nodes(G_named, pos, node_size=2500)
nx.draw_networkx_edges(G_named, pos, width=1.2, alpha=0.8)
nx.draw_networkx_labels(G_named, pos, font_size=10)

plt.title("Krackhardt Kite Graph")
plt.axis("off")
plt.show()
```
